```python
import json

with open('./wordnet2.json', 'r') as file_word:
    input_file = file_word.read()

dict_wordnet = json.loads(input_file)
```

```python
list(dict_wordnet)
new_dict = {"wordnet": []}
for header in list(dict_wordnet):
    temp = {
        "input": dict_wordnet[header]['sinonim'],
        "output": header
    }
    new_dict['wordnet'].append(temp)
```
